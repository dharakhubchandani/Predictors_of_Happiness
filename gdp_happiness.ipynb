**Question 2: What are the Best Predictors of Happiness in a country? Is our model significant?**

```python
import statsmodels.api as sm
import itertools

final_df['Perceptions of Corruption'].fillna(0, inplace=True)

# Defining the x variables and y variables
x_variables = final_df[['GDP per Capita', 'Family','Life Expectancy','Freedom', 'Perceptions of Corruption', 'Generosity','GINI Coefficient']]
y_variables = final_df['Happiness Score']

# Stepwise regression
def stepwise_regression(X, y):
    remaining_features = list(X.columns)
    best_score = -float('inf')
    best_model = None
    
    for k in range(1, len(X.columns) + 1):
        for subset in itertools.combinations(remaining_features, k):
            model = sm.OLS(y, X[list(subset)]).fit()
            score = model.rsquared_adj
            
            if score > best_score:
                best_score = score
                best_model = model
                best_subset = subset
                
        remaining_features = list(set(remaining_features) - set(best_subset))
        
    return best_model

model = stepwise_regression(x_variables, y_variables)

print(model.summary())

print("Regression Coefficients:")
for i, feature in enumerate(model.params.index):
    if i == 0:
        print(f"Intercept: {model.params[i]}")
    else:
        print(f"{feature}: {model.params[i]}")
        
print("\nP-values:")
print('BASED ON STEPWISE REGRESSION')
print(model.pvalues)

print(" ")
print("The 2 variables which are significant in predicting the cholesterol levels are 'weight' and 'Glucose levels'")        
```

**Question 3: Do Countries with higher GDPs have higher happiness score?**

```python
#Do countries with higher GDPs have higher happiness score?
import matplotlib.pyplot as plt

# create a list of years to iterate through
years = [2017, 2018, 2019]

# create a figure with 2 rows and 5 columns
fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(20, 10))

# loop through each year and create a bar chart for top GDP countries
for i, year in enumerate(years):
    top_gdp = final_df[final_df['Year'] == year].sort_values(by='GDP', ascending=False).head(5)
    ax[0, i].bar(top_gdp.index, top_gdp['GDP'])
    ax[0, i].set_title(f'Top GDP Countries in {year}')
        

# loop through each year and create a bar chart for top Happiness Score countries
for i, year in enumerate(years):
    top_happiness = final_df[final_df['Year'] == year].sort_values(by='Happiness Score', ascending=False).head(5)
    ax[1, i].bar(top_happiness.index, top_happiness['Happiness Score'])
    ax[1, i].set_title(f'Top Happiness Score Countries in {year}')
    ax[1, i].set_ylim([6,8])

# adjust the spacing between subplots
plt.subplots_adjust(hspace=0.5, wspace=0.2)

# display the plot
plt.show()

print("We can analyse that the countries which are top GDP countries are not the top happy countries.")
```
